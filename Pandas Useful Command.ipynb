# Making Changes to the Data

```python
#df.head(5)

##add new column with total addtion of row values

#df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
#df.head(5)

##easier way to do same ... " : " means all
df['Total'] = df.iloc[:,4:10].sum(axis=1)
df.head(5)
```

```python
#ALL THESE COMMAND DON'T REALLY MODIFY DATA AT ALL.. IS HELP VISUALIZE THE CHANGES

#AVOID HARDCOADING THE VALUES OF COLUMN AS BELOW ..BETTER TRY TO FETCH INDEX OF PARTICULAR COLUMN AND USE

# moving total column from one place to another
# cols[0:4] will get first 4 column in same order we want it
#cols[-1] reverse indexing to get the total 
#cols[4:12] 12 is the position  of the column need to be moved after 4th column
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

# Saving  our Data (Exporting into Desired Format) 

```python
#SAVING DATA IN CSV FORMAT where we don't want index
df.to_csv('modified .csv',index=False)
```

```python
#SAVING DATA IN excel FORMAT where we don't want index
df.to_excel('modified.xlsx', index=False)
```

```python
#SAVING DATA IN txt FORMAT or TAB seperated format where we don't want index with seperator 
df.to_csv('modified.txt',index=False, sep='\t' )
```

# Filtering Data

```python
# one filtering condition
df.loc[df['Type 1']== 'Grass']
```

```python
# Two filtering condition
df.loc[(df['Type 1']== 'Grass') & (df['Type 2']== 'Poison')]
```

```python
# Multiple filtering condition
df.loc[(df['Type 1']== 'Grass') & (df['Type 2']== 'Poison') & (df['HP']>70) ]
```

```python
# To reset index for the data frame (df)

new_df= df.loc[(df['Type 1']== 'Grass') & (df['Type 2']== 'Poison') & (df['HP']>70) ]
new_df= new_df.reset_index()
new_df
```

```python
# when you want to search in a column which contain some "specific" words

df.loc[df['Name'].str.contains('Mega')]
```

```python
#RegEx can be used to check if a string contains the specified search pattern.
# to ignore uppercase issue we can use FLAG "flags=re.I"
import re
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I, regex= True)]
```

```python
# * means zero or more  and ^ means it word should start with pi 
import re
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I, regex= True)]
```

# Conditional Changes

```python
df.loc[df['Type 1']== 'Fire', 'Legendary']= True
df
```
